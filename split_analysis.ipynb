import numpy as np
import cv2
import os
import pandas as pd
from IPython.display import Image
import matplotlib.pyplot as plt

```python
folder_name = 'split-sample-vid'

images = [ folder_name+'/'+file_name for file_name in os.listdir(folder_name)]

arr = []
```

```python
for i in range(len(images)-1):
  pair = images[i:i+2]
  a = cv2.imread(pair[0])
  b = cv2.imread(pair[1])
  n = (np.linalg.norm(b-a))*10000/(a.shape[0]*a.shape[1])
  element = {'norm':n,'file_1':pair[0],'file_2':pair[1]}
  arr.append(element)
```

```python
df = pd.DataFrame(arr)
df.sort(columns="norm",ascending=False)
```
